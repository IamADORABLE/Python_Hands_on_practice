```python
def print_board(b):
    print("\nBoard:")
    for r in b:
        print(" ".join(str(n) if n else "." for n in r))

def is_valid(b, r, c, n):
    return n not in b[r] and all(b[i][c] != n for i in range(5))

def play():
    board = [
    [1, 0, 5, 0, 3],
    [0, 4, 1, 0, 2],
    [0, 0, 0, 1, 0],
    [4, 0, 0, 2, 0],
    [0, 1, 0, 0, 5]
]


    while any(0 in row for row in board):
        print_board(board)
        try:
            r = int(input("Row (0-4): "))
            c = int(input("Col (0-4): "))
            n = int(input("Num (1-5): "))
            if board[r][c] != 0:
                print("‚ùå Filled.")
            elif is_valid(board, r, c, n):
                board[r][c] = n
            else:
                print("‚ùå Invalid.")
        except:
            print("‚ö†Ô∏è Try again.")

    print_board(board)
    print("üéâ Done!")

play()
```
